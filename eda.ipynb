Thus, we can also see a rather significant increase in the danceability of the most popular songs from one period to another

To better understand the variance from one period to another, we can do a KDE plot


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.kdeplot(top_a['danceability'], label='Top Hits 2000-2015', fill=True)
sns.kdeplot(top_b['danceability'], label='Top Hits 2016-2023', fill=True)
plt.title('Distribution of Danceability for Top 10% Popular Songs')
plt.legend()
plt.show()
```

We can clearly see a shit on danceability to the right, showcasing that for the most popular songs, danceability has significantly increased.
More than that, we can also see that the orange slope, for Period B, has become also thinner - what this suggests, is that on Period A, there was more space for songs with lower danceability to be popular, while on Period B onwards, this less true

Now, what we can finally do, is a T-test, to understand if this finding is statistically significant.

```python
t_stat, p_val = stats.ttest_ind(top_a['danceability'], top_b['danceability'])
```

```python
print(f"P-value: {p_val}")
if p_val < 0.05:
    print("The shift in danceability is statistically significant!")
```

Now, we have done this comparison between the two periods, considering the most popular songs.

However, it is still not enough to conclude, that popularity is dependent on danceability

With the following methodology, we can understand the probability of a songs being both danceable within both populations

```python
#First, we set a threshold of danceability - lets say 0.7
threshold = 0.7
```

```python
high_dance_hits = len(top_b[top_b['danceability'] > threshold])
high_dance_flops = len(bottom_b[bottom_b['danceability'] > threshold])
```

```python
if high_dance_flops > 0:
    ratio = high_dance_hits / high_dance_flops
    print(f"In Period B, a song with > {threshold} danceability is {ratio:.2f}x more likely to be a Hit than a Flop.")
else:
    print("No flops found in this range, suggesting danceability is an incredibly strong filter.")
```

### Conclusion: 
Since 2016 (Period B), danceability has not only increased across the board but has also become a key separator for success. While a high danceability score is now a "baseline" for most tracks, songs exceeding a 0.7 danceability threshold are 1.53x more likely to be in the Top 10% of popularity compared to the Bottom 10%, confirming its role as a significant predictor of modern hit-making.

### 3.1 Objective and Labelling

```python
# 3.1.1 Defining the Threshold
# First we need to define the threshold of what it takes to be "a hit"
threshold = df['popularity'].quantile(0.75)
```

```python
# 3.1.2 Binary Classification
# Then we need to create a binary column: 1 for Hit, 0 for Not a Hit.
df['is_hit'] = (df['popularity'] >= threshold).astype(int)
```

### 3.2 Features Selection

```python
# 3.2.1 Numerical Isolation
# Now we need to define, out of all variables in this study, which ones matter - In this case, we are keeping only numerical values
features = ['year', 'danceability', 'energy', 'key', 'loudness', 'mode', 
            'speechiness', 'acousticness', 'instrumentalness', 'liveness', 
            'valence', 'tempo', 'duration_ms', 'time_signature']
```

```python
# 3.2.2 Separating X and Y, separating the features from the result (is the song a hit or a flop?)
X = df[features]
y = df['is_hit']
```

### 3.3 The Temporal Split

```python
# 3.3.1 Training
# First the model needs to "study" the relationship between X and Y in the first 20 years of the study
X_train = X[X['year'] <= 2020]
y_train = y[X['year'] <= 2020]
#3.3.2 Testings
# Now the model needs to the same, but for the 3 last years of the study, from 2021-23
X_test = X[X['year'] > 2020]
y_test = y[X['year'] > 2020]
```

### 3.4 Setting the Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# 3.4.1 N-Estimators
# First thing we need to do, is define how many estimators will be used in this study - Thus we defined 100 "independent decision trees", each looking at a randon subset of songs and features
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)


# 3.4.2 "Voting" System
# Running the model.predict we all 100 trees "vote." If 70 trees say a song is a hit based on its loudness and duration, the model outputs a "1." This "averaging" makes your model much more stable than a single human observation.
predictions = model.predict(X_test)


print(f"Model Accuracy on 2021-2023: {accuracy_score(y_test, predictions):.2f}")
print(classification_report(y_test, predictions))
```

* Strength (The Pattern exists): The model achieved 67% accuracy, which is significantly higher than a random guess. This proves that a song’s "DNA" (audio features) still plays a dominant role in its success, even in the modern era.

* Limitation: While the model is great at identifying "Flops," its recall for Hits is only 54%. This means the model—trained on historical data—is missing nearly half of today’s successes. It’s essentially a coin flip when trying to catch a modern hit, suggesting the "rules" have shifted.*

```python
from sklearn.metrics import confusion_matrix
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Get feature importance from the model
importances = model.feature_importances_
feature_names = features # This is the list we defined earlier

# Create a DataFrame for easy plotting
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color='skyblue')
plt.gca().invert_yaxis()
plt.title('Which Features Predict a Hit?')
plt.xlabel('Importance Score')
plt.show()
```

```python
# 1. Identify the 'Predicted Hits' from your model's results
# We use the X_test (2021-2023) and the predictions we just made
X_test_with_preds = X_test.copy()
X_test_with_preds['predicted_hit'] = predictions
X_test_with_preds['actual_hit'] = y_test

# 2. Calculate the 'Expected' profile (Average of Predicted Hits)
expected_profile = X_test_with_preds[X_test_with_preds['predicted_hit'] == 1][features].mean()

# 3. Calculate the 'Actual' profile (Average of Actual Hits)
actual_profile = X_test_with_preds[X_test_with_preds['actual_hit'] == 1][features].mean()

# 4. Create a Comparison DataFrame
comparison_df = pd.DataFrame({
    'Expected (Model)': expected_profile,
    'Actual (Reality)': actual_profile
})

# Add a 'Difference' column to see the gap
comparison_df['Difference %'] = ((comparison_df['Actual (Reality)'] - comparison_df['Expected (Model)']) / comparison_df['Expected (Model)']) * 100

print("--- 2021-2023 Hit Formula: Model vs. Reality ---")
print(comparison_df.round(2))
```
